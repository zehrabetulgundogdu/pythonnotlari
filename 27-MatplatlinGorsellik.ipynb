```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
numpyDizisi1= np.linspace(0,10,20)
```

```python
numpyDizisi2= numpyDizisi1**2
```

```python
(benimFigur,benimEksen) = plt.subplots()
```

```python
(benimFigur,benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"g")
benimEksen.plot(numpyDizisi2,numpyDizisi1,"r")
```

```python
(yeniFigur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi1+2,color="blue")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+3,color="yellow")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+4,color="r",linewidth=5.0,linestyle="--")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+5,color="g",linewidth=4.0, linestyle=":")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+6,color="#000000",linewidth=6.0 , linestyle="-.")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+6,color="#43021b",linewidth=2.0 , linestyle="-.",marker="o",markersize=5,markerfacecolor="r")
```

```python
#linewidth= çizgi kalınlığı
#linestyle= çizgistili
#marker= değerin geldiği noktayı belirtmek için kullanılır.
#markersize=boyutu
```
